### Step 1 : Create source embeddings for the text column

```python
from sentence_transformers import SentenceTransformer
```

```python
encoder = SentenceTransformer("all-mpnet-base-v2")
vectors = encoder.encode(df.text)
```

```python
vectors.shape
```

```python
dim = vectors.shape[1]
dim
```
